```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('C:/Users/Manhab Zafar/Desktop/CNET-Iteration-2/data/raw/InSDN_Normal_and_Attack_Combined.csv')

# Basic information
print("Dataset Shape:", df.shape)
print("\nColumn Names:\n", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nClass Distribution:\n", df['Label'].value_counts())

# Statistical summary
print("\nStatistical Summary:\n", df.describe())

# Visualize class distribution
plt.figure(figsize=(8, 6))
df['Label'].value_counts().plot(kind='bar')
plt.title('Class Distribution')
plt.xlabel('Class (0: Normal, 1: Attack)')
plt.ylabel('Count')
plt.savefig('C:/Users/Manhab Zafar/Desktop/CNET-Iteration-2/results/figures/class_distribution.png')
plt.show()

# Check for attack types
if 'attack_type' in df.columns:
    print("\nAttack Types:\n", df['attack_type'].value_counts())
    
    plt.figure(figsize=(10, 6))
    df['attack_type'].value_counts().plot(kind='bar')
    plt.title('Attack Type Distribution')
    plt.xlabel('Attack Type')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('C:/Users/Manhab Zafar/Desktop/CNET-Iteration-2/results/figures/attack_types.png')
    plt.show()
```
